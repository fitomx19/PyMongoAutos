Archivo Generador de Nombres Aletorios

```python
import csv
NombresEnBruto = []

with open('nombres.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        
        NombresEnBruto.append(row)
```

```python
Nombre = []
for nombre in NombresEnBruto:
        Nombre.append((nombre[0].split(" ", 1)))

nombre = [item[0] for item in Nombre]

apellido = [item[1] for item in Nombre]

```

```python
import datetime,random
FechaNacimiento = []
FechaNacimientoRFC = []

for y in range(100):
    y = datetime.datetime((random.randint(1970, 2000)), (random.randint(1, 12)), (random.randint(1, 28)))
    print(y.strftime("%d/%m/%Y"))
    FechaNacimiento.append(y.strftime("%d/%m/%Y"))
    FechaNacimientoRFC.append(y.strftime("%d%m%Y"))

```

PseudoRFC
Vamos a Generar una especie de RFC para evitarnos todo el estricto rollo

* Las primeras dos letras (VE) son el apellido paterno más la primera vocal interna del apellido paterno.
* El tercer dígito (C) es la inicial del apellido materno. De no existir un apellido materno se utiliza una (X).
* El cuarto dígito (J) es la inicial del primer nombre.
* Los primeros dos dígitos son el año de nacimiento (88).
* Los segundos dígitos son el mes de nacimiento (03 o marzo).
* Los terceros dígitos son el día de nacimiento (26).
* Por lo tanto la persona nació el veintiséis de marzo de 1988.
* Los últimos dígitos (XXX) se le conoce como homoclave, esta la designa el SAT, dependiendo de algunos factores que realiza el SAT por medio de sistemas numéricos o alfanuméricos.

```python
RFCs = []

for x in range(99):
    rfc = ""
    #primer letra nombre
    rfc = (nombre[x][0])
    #Obtenemos apellido Materno/Paterno
    Apellido = (apellido[x].split(" " ,2))
    #Imprimimos Las dos letras upper del Apellido Paterno
    rfc = rfc + (Apellido[0][0:2]).upper()
    #Checamos si tiene apellido Materno , si no usamos la letra 3 del paterno
    try:
       rfc = rfc +  (Apellido[1][0]).upper()
    except IndexError:
        rfc = rfc + (Apellido[0][3]).upper()

    #fecha de nacimiento en numeros seguidos
    rfc = rfc + (FechaNacimientoRFC[x])

    rfc = rfc + str(random.randint(100, 999))

    RFCs.append(rfc)

    
```

Generador de Vehiculos

```python
Vehiculos = []

with open('vehiculos.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        
        Vehiculos.append(row)

#["'2017", " Chevrolet Spark': '$ 169", '000\\nMensualidad $ 3', "288'", '']
```

```python
NombreVehiculo = []
for x in range(len(Vehiculos)-18):


    y = Vehiculos[x][1].split(":" , 1)[0].replace('"', '')
    y = y.translate(str.maketrans({"'":None}))
    NombreVehiculo.append(y)
NombreVehiculo
```

```python
AnioAuto = []
for x in range(len(Vehiculos)-18):
    y = Vehiculos[x][0].split(":" , 1)[0].replace('"', '')
    y = y.translate(str.maketrans({"'":None}))
    AnioAuto.append(y)
AnioAuto
```

```python
Renta = []
for x in range(len(Vehiculos)-18):
    y = (Vehiculos[x][3].split(":" , 1)[0].replace('"', ''))
    y = y.translate(str.maketrans({"'":None}))
    Renta.append(y)
Renta
```

```python
Color = []
Combustible = []
KM = []
Colores = ["Rojo","Blanco","Negro","Gris","Azul","Plateado","Verde","Amarillo"]
Combustibles = ["Gasolina","Diesel"]
for x in range(len(Vehiculos)-18):
    Color.append((Colores[random.randint(0,7)]))
    Combustible.append((Combustibles[random.randint(0,1)]))
    KM.append((random.randint(10000,100000)))
```

```python
import string

PLACAS = []
number_of_strings = len(Vehiculos)-18
length_of_string = 3
for x in range(number_of_strings):
  y = str(random.randint(100,999)) + "-" +''.join(random.choice(string.ascii_letters + string.digits) for _ in range(length_of_string))
  y = y.upper()
  PLACAS.append(y)
```

```python

df4 = pd.DataFrame(list(zip(NombreVehiculo, AnioAuto,Renta,Color,KM,Combustible,PLACAS)),
               columns =['Auto', 'Modelo', 'Renta','Color','Kilometraje','Combustible','Placa'])
```
